```python
from bs4 import BeautifulSoup
```

```python
with open ("scrap_data/page1.html",  "r") as f:
    html_content=f.read();
```

```python
soup = BeautifulSoup(html_content, "lxml");
```

```python
all_h3 = soup.find_all("h3");
```

```python
all_country_info = []
for h3 in all_h3:
    name=h3.get_text(strip = True)
    population = h3.find_next("div").select("span.country-population")[0].get_text(strip = True)
    all_country_info.append([name,population])
```

```python
import pandas as pd

df=pd.DataFrame(all_country_info, columns = ["name", "population"])
```

```python
df.to_csv("clean_data/cleaning_data.csv",index = False)
```
