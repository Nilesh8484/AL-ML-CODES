```python
import json
```

```python
def data_load(fileName):
    with open(fileName,"r") as f:
        data = json.load(f);
    return data;
```

```python
data=data_load("store_data.json");
```

```python
print(data);
```

```python
def clean_data(data):
    uniq_set= set()
    clean_datas=[]

    #mix data problem
    text_to_num = {"one":1,"two":2, "three":3, "four":4,"five":5};
    for user in data:
        raw_rating= str(user.get("rating")).strip().lower()
        if (raw_rating in text_to_num):
            raw_rating= text_to_num[raw_rating]
        
        user["rating"]=raw_rating;   

        
        #miss data 
        raw_age = user.get("age");
        if (raw_age == None):
            user["age"]=None;


        if (user["name"].strip() in uniq_set):
            continue
        uniq_set.add(user["name"]);
        clean_datas.append(user)

    return clean_datas
            
```

```python
data=clean_data(data)
```

```python
def get_insight(data):
    tot_price = 0;
    for user in data:
        tot_price +=float(user["rating"]);
    print(f"avg rating is = {tot_price/len(data)}")

    poor_rating=0
    for user in data:
        if (float(user.get("rating")) < 3):
            poor_rating += 1
    print(f"number of poor rating is= {poor_rating/len(data)*100}")
        
```

```python
get_insight(data);
```

```python
# make recommendation 

def get_recommedation(data):
    recommadation = []
    
    for user in data:
        current = {}
        current["name"] = user["name"];
        if (float(user.get("rating")) >= 4):
            current["brand"] = "apple"
        else:
            current["brand"] = "samsung"
        recommadation.append(current);

    return recommadation
            
```

```python
get_recommedation(data)
```
