```python
tips=sns.load_dataset("tips")

print(tips.head())



sns.barplot(
    data= tips,
    x= "day",
    y = "tip",
    hue= "sex"
)
```

```python
penguins = sns. load_dataset("penguins")

sns.histplot(
    data= penguins,
    x= "body_mass_g",
    bins = 30,
    color = "red"
    
)

# penguins.head()
```

```python
flight_pivot=flight.pivot(index= "month",columns="year", values= "passengers")

sns.heatmap(
    flight_pivot,
    cmap= "coolwarm",
    fmt= "d",
    annot = True
    



)
plt.title("heatmap for flight")
plt.tight_layout()
```
