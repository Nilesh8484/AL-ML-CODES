```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import datasets
from sklearn.metrics import r2_score
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVR
```

```python
df= datasets.load_diabetes(as_frame=True).frame
```

```python
df.head()
```

```python
X = df.drop("target", axis=1)
y=df["target"]
```

```python
X_train,X_test,y_train,y_test= train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
y_scale = StandardScaler()

y_train_scale= y_scale.fit_transform(y_train.values.reshape(-1,1)).ravel()
y_test_scale= y_scale.transform(y_test.values.reshape(-1,1)).ravel()
```

```python
model= SVR()
model.fit(X_train, y_train_scale)
y_pred_scale=model.predict(X_test)
print("r2_score", r2_score(y_test_scale,y_pred_scale))
```

```python
model= SVR(kernel="linear")
model.fit(X_train, y_train_scale)
y_pred_scale=model.predict(X_test)
print("r2_score", r2_score(y_test_scale,y_pred_scale))
```

```python
model= SVR(kernel="poly")
model.fit(X_train, y_train_scale)
y_pred_scale=model.predict(X_test)
print("r2_score", r2_score(y_test_scale,y_pred_scale))
```

```python
model= SVR(kernel= "sigmoid")
model.fit(X_train, y_train_scale)
y_pred_scale=model.predict(X_test)
print("r2_score", r2_score(y_test_scale,y_pred_scale))
```

```python
from sklearn.model_selection import GridSearchCV

param_grid = {
    "C":[1,2,3,4,5,10],
    "kernel":["liner", "rbf"],
    "epsilon":[0.1,0.2,0.3]
}
```

```python
svr =SVR()

grid_search = GridSearchCV(svr, param_grid, scoring="r2", cv=5, max_iter=7000)

grid_search.fit(X_train, y_train_scale)
```

```python
print(grid_search.best_params_)
```

```python
best_model= SVR(C= 1, epsilon=0.4, kernel= "rbf")
best_model.fit(X_train, y_train_scale)
y_pred_scale=best_model.predict(X_test)
print("r2_score", r2_score(y_test_scale,y_pred_scale))
```
