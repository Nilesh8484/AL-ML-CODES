```python
# Uniform Distribution
import numpy as np
import matplotlib.pyplot as plt

# values from 0 to 10
values = np.random.uniform(0, 10, 10_00_000)

# plotting a histogram - becomes more uniform with more values
plt.hist(values, bins=100, density=True, alpha = 0.3,edgecolor = "black")

plt.title("Continuous Uniform Distribution between 0 and 10")
plt.xlabel("x")
plt.ylabel("Probability Density")
plt.grid()

plt.show()
```

```python
# Normal Distribution
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parameters
mu = 70
sigma = 10

# X-axis values
x = np.linspace(30, 110, 1000)

# mean
# standard deviation
# Normal PDF
y = norm.pdf(x, mu, sigma)

# Plot the bell curve
plt.plot(x, y, color='black', linewidth=2, label='Normal Distribution')

# Shade 68% region (p ± lo)
x1 = np. linspace(mu - sigma, mu + sigma, 1000)
plt.fill_between(x1, norm.pdf(x1, mu, sigma), color='green', alpha=0.3, label='68% region (1g)')


# Shade 95% region (u ± 2g)
x2 = np. linspace(mu - 2*sigma, mu + 2*sigma, 1000)
plt.fill_between(x2, norm.pdf(x2, mu, sigma), color='yellow', alpha=0.2, label='95% region (2g)')

# Shade 99.7% region (p ± 3g)
x3 = np. linspace(mu - 3*sigma, mu + 3*sigma, 1000)
plt.fill_between(x3, norm.pdf(x3, mu, sigma), color='red', alpha=0.1, label='99.7% region (3g)')

# Labels and title
plt.title("Normal Distribution of Student Scores")
plt.xlabel("Score")
plt.ylabel("Probability Density")
plt. legend()
plt. show()
```

```python
# Plot y = Vx
import numpy as np
import matplotlib.pyplot as plt

# Define x values (x >= 0 for sqrt)
x = np.linspace(0, 100, 100_000)
y = np.sqrt(x)

# Plot
plt.plot(x, y, label="y = Vx")
plt.xlabel("x")
plt.ylabel("y")
plt.title("y = Vx")
plt. legend()
plt.grid(True)

plt.show()
```
